```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Types of plots

#### line plots

```python
x = [1,2,3,4,5]
y = [10, 20, 35, 40, 70]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel('X-axis');
plt.ylabel('Y-axis');
plt.grid(True)
```

#### Bar Charts

```python
months = ['jan','feb','march','apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue);
plt.title('month revenue bar charts');
plt.xlabel('Month');
plt.ylabel('Revenue');
```

#### Scatter Plots

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 167, 145]
```

```python
plt.scatter(heights, weights);
plt.title('Height vs Weight scatter plots')
plt.xlabel('Height (inches)');
plt.ylabel('Weight (pounds)');
```

#### subplots & legends

```python
days = [1,2,3,4,5]
original_price = [11,17,20,24,30]
predicted_price = [9,14,20,23,29]
```

```python
plt.plot(days, original_price,'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.xlabel('weekdays');
plt.ylabel('Price');
```

#### subplots

```python
# first plots
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original Price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

# second plots
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted Price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()


# Third plots
price_diff = [o-p for o,p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_diff, color='purple', label='Price diff')
plt.title('diff between Original & Predicted Price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price diff')
plt.legend()

plt.tight_layout()
```

### saving the fig

```python
plt.plot(days, original_price,'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.legend();
plt.xlabel('weekdays');
plt.ylabel('Price');
plt.savefig('example_prediction.png')
```
